```python
# Reading the base config file
BASE_CONFIG_FILE = os.path.join(root_folder, "base_config.yaml")
with open(BASE_CONFIG_FILE, "r") as file:
    base_config_dict = yaml.safe_load(file)


# Specify which networks to read
## data_folder_name is the same as variable in `path_configs` in config.yaml
data_folder_name = base_config_dict["path_configs"]["data_folder_name"]

## project_name is the same as variable in `path_configs` in config.yaml
project_name = base_config_dict["path_configs"]["project_name"]

## input_scenario_name is the same as variable in `path_configs` in config.yaml
input_scenario = base_config_dict["path_configs"]["input_scenario_name"]

## output_scenario_name is the same as variable in `path_configs` in config.yaml
scenario = base_config_dict["path_configs"]["output_scenario_name"]

## sector is the same as variable in `base_configs` in config.yaml, only the first element is read
sector = base_config_dict["base_configs"]["sector"][0]

## years is the same as variable in `base_configs` in config.yaml, all years are read
years = base_config_dict["base_configs"]["years"]

## either `pre-solve`, `pre-solve-brownfield` or `post-solve`
network_stage = "post-solve"
```

```python
# Reading the scenario config file
SCENARIO_CONFIG_PATH = os.path.join(
    root_folder, "data", data_folder_name, project_name, "input", input_scenario
)

yaml_files = glob.glob(f"{SCENARIO_CONFIG_PATH}/*.yaml")
with open(yaml_files[0], "r") as f:
    scenario_configs_dict = yaml.safe_load(f)

# Specify which secnario settings to read
if scenario_configs_dict["scenario_configs"]["resolution"]["method"] == "nth_hour":
    nth_hour = scenario_configs_dict["scenario_configs"]["resolution"]["stepsize"]
else:  # clustered method
    nth_hour = 1
```

```python
# Get list of networks
networks = [
    f"data/{data_folder_name}/{project_name}/results/{scenario}/{network_stage}/network_{sector}_{x}.nc"
    for x in years
]
network_paths = [os.path.join(root_folder, x) for x in networks]
csvs = [
    f"data/{data_folder_name}/{project_name}/results/{scenario}/csvs/{sector}/{str(x)}/"
    for x in years
]
csvs_paths = [os.path.join(root_folder, x) for x in csvs]

# load output table classes
ot = OutputTables(
    network_list=network_paths,
    config=base_config_dict,
    scenario_configs=scenario_configs_dict,
)
```
